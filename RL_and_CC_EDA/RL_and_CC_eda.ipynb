```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Prep of The Datasets

```python
## Prep of the Rule of Law Data ##

filter = ['Country Name','2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
 '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017',
 '2018', '2019', '2020']
rl_df = pd.read_csv("API_RL.PER.RNK_DS2_en_csv_v2_27865.csv")
rl = rl_df[filter]
rl.head()
```

```python
## Prep of the Control of Corruption Data ##

cc_df = pd.read_csv("API_CC.PER.RNK_DS2_en_csv_v2_21140.csv")
cc = cc_df[filter]
cc.head()
```

### Analysis of Turkey

```python
rl_turkey = rl[rl['Country Name'] == 'Turkiye']
cc_turkey = cc[cc['Country Name'] == 'Turkiye']
```

```python
years = rl_turkey.columns[1:].astype(int)
years
```

```python
rl_values = rl_turkey.values[:,1:][0]
rl_values[1] = (rl_values[0] + rl_values[2]) / 2 #imputing the nan value with the avarage of the previous and the next values
rl_values
```

```python
cc_values = cc_turkey.values[:,1:][0]
cc_values[1] = (cc_values[0] + cc_values[2]) / 2 #imputing the nan value with the avarage of the previous and the next values
cc_values
```

```python
plt.Figure(figsize=(10,8))
plt.title("Rule of Law and Control of Corruption percentile Rank of Turiye over the Years")
plt.plot(years, rl_values, label='Rule of Law % Rank',color='teal')
plt.xticks(years, rotation=90)

plt.plot(years, cc_values, label= 'control of Corruption % Rank', color='orange')


plt.legend()
plt.show()
```
